# Data Validation with Great Expectation

```python
# Import Libraries

import pandas as pd
import great_expectations as gx
import warnings
warnings.filterwarnings('ignore')
```

```python
# Get dataframe

df = pd.read_csv("P2M3_Moses_Imanuel_Salim_data_clean.csv")
```

```python
# Create Data Context

context = gx.get_context()
```

```python
# Connect to data and create a Batch

data_source = context.data_sources.add_pandas("pandas")
data_asset = data_source.add_dataframe_asset(name="pd dataframe asset")

batch_definition = data_asset.add_batch_definition_whole_dataframe("batch definition")
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})
```

```python
# Creat an expectation suite and validator
expectation_suite_name = 'expectation-employee-dataset'

validator = context.get_validator(
    batch=batch,
    create_expectation_suite_with_name="expectation-employee-dataset"
)
```

```python
# Validation to be unique

validator.expect_column_values_to_be_unique("employeenumber")
```

Untuk validation ini dilakukan pengecekan untuk keunikan value employee number. Pada dataset ini Employee Number dapat dibilang seperti ID employeenya. Karena itu jika ada Employee ID yang sama, kemungkinan adanya baris yang duplikat. Tetapi dari hasil validasi tersebut terlihat kalau tidak ada Employee Number yang sama hingga tidak ada baris yang duplikat

```python
# Validation between x and y

validator.expect_column_values_to_be_between(
    column = "age",
    min_value=18,
    max_value=60,
)
```

Validation ini melakukan cek apakah umur dari employee berada antara 18 hingga 60. Pada normalnya, hanya yang sudah berumur 18 keatas yang dapat bekerja di perusahaan. Dan juga jika umur sudah di atas 60 maka pegawai tersebut harusnya sudah pensiun. Karena itu dilakukan cek to be between 18 sebagai min value dan 60 sebagai max value.

```python
# Validation to be in set

validator.expect_column_values_to_be_in_set(
    "attrition",
    ["Yes", "No"]
)
```

Validation ini dilakukan pada kolom attrition dimana value yang harusnya ada pada kolom ini hanya pada yes and no. Kolom ini membicarakan apakah employee tersebut keluar dari perusahaan atau tidak karena itu isi dari kolom ini tidak mungkin diluar dari yes and no

```python
# Validation to be in type list

validator.expect_column_values_to_be_in_type_list(
    "monthlyincome",
    ["int", "float"]
)
```

Validation ini dilakukan untuk cek apakah datatype dari kolom tersebut sudah benar. Dilakukan cek apakah kolom monthly income sudah berupa integer atau float. Unutk dataset seperti ini terkadang kolom yang berhubungan dengan uang mungkin masih berupa string karena terdapat mata uangnya. Karena itu dilakukan pencekan apakah kolom ini sudah bersih dan berbentuk integer atau float

```python
# Validation most common value to be in set

validator.expect_column_most_common_value_to_be_in_set(
    "StandardHours",
    [80]
)
```

Validation ini dilakukan untuk mengetahui apakah value paling sering muncul pada sebuah kolom adalah 80. Standard hours untuk employee bekerja adalah 80. Dilakukan cek apakah pada dataset ini kolom standard hours nya sudah mengikuti standard

```python
# Validation column pair values A to be greater than B

validator.expect_column_pair_values_A_to_be_greater_than_B(
    "YearsAtCompany",
    "YearsInCurrentRole",
    or_equal=True
)
```

Validation ini untuk mengetahui apakah jumlah dari kolom A lebih besar sama dengan dari kolom B. Pada kali ini 2 kolom yang dilakukan pengencekan untuk kolom years at company dan years in current role. Tahun bekerjanya di perusahaan tersebut tidak mungkin lebih kecil dari tahun pada role tersebut. Namun, jumlahnya sama masih mungkin karena mungkin employee tersebut tidak pernah pindah role dari awal masuk perusahaan. Karena itu pengecekannya adalah kolom A lebih besar sama dengan kolom B.

```python
# Validation column unique value count to be between

validator.expect_column_unique_value_count_to_be_between(
    "Gender",
    min_value=1,
    max_value=3
)
```

Validation ini dilakukan untuk mengecek unique value count dari kolom gender. Pada dataset employee normalnya gender biasa hanya terdiri dari 2 atau 3. Antara Male, Female, dan terkadang ada unknown atau others. Untuk kolom unknown atau others masih mungkin karena mungkin terdapat missing value atau tidak ingin memberi tahu gender mereka. Namun, jika unique value dari kolom gender sudah melebihi 3 maka sudah tidak masuk akal. Karena itu dilakukan cek apakah jumlah unique valuenya tidak melebihi 3.
